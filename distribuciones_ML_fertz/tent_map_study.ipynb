```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
mu = 1.5  # Parámetro de control para el mapeo
n = 50  # Número de iteraciones
x0 = 0.2  # Valor inicial de x

# Inicializamos el array para almacenar los valores
x_values = np.zeros(n)
x_values[0] = x0

# Aplicamos el mapeo "tienda de campaña"
for i in range(1, n):
    if x_values[i-1] <= 0.5:
        x_values[i] = mu * x_values[i-1]
    else:
        x_values[i] = mu * (1 - x_values[i-1])

# Graficamos los puntos obtenidos
plt.figure(figsize=(10, 6))
plt.plot(range(n), x_values, 'bo-', markersize=5)
plt.title('Mapeo "Tienda de Campaña" - Comportamiento Ordenado (μ = 1.5)')
plt.xlabel('Número de Iteración')
plt.ylabel('Valor de x')
plt.grid(True)
plt.show()
```

# medidas de dispersión

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros para la población
X0 = 0.4  # Población inicial (puede estar entre 0 y 1)
r = 4  # Parámetro de crecimiento
n = 101  # Número de iteraciones

# Inicializamos el array para almacenar los valores de población
population = np.zeros(n)
population[0] = X0

# Simulación de la población
for i in range(1, n):
    population[i] = (r * population[i-1]) - r*(population[i-1])**2
    
# guardamos los valores de 'population' en un objeto 'np.array'
population = np.array(population)

# calculamos la media
mean_population = np.mean(population)
print(f"La media de la población es: {mean_population}")
# calcular la Varianza
var_population = np.var(population)
print(f"La varianza de la población es: {var_population}")
# Se puede obtener la Asimetría estadística?
skew_population = np.mean((population - mean_population)**3) / var_population**(3/2)
print(f"La asimetría estadística de la población es: {skew_population}")
# Graficar histograma de frecuencias
plt.figure(figsize=(10, 6))
plt.hist(population, bins=20, edgecolor='black')
plt.title('Histograma de Frecuencias de la Población')
plt.xlabel('Población')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
