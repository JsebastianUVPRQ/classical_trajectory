Histograma de densidades

```python
# cargar archivo csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

# import statsmodels.api as sm
# import statsmodels.formula.api as smf
# from statsmodels.stats.outliers_influence import variance_inflation_factor
# from statsmodels.stats.proportion import proportions_ztest
```

```python
# carga el archivo csv
df = pd.read_csv('datos-densidades.csv', header=0)

# muestra las primeras 5 filas del dataframe
df.head()
```

```python
# histograma con los nuevos headers 
# Grupo,PesoNa2CO3,VolumenTitNa2CO3,VolumenTitNaOH,[HCl]mol/L,[NaOH] mol/L

df_final = pd.read_csv('datos-densidades.csv', header=0)

df_final.columns = ['Grupo', 'PesoNa2CO3', 'VolumenTitNa2CO3', 'VolumenTitNaOH', '[HCl] mol/L', '[NaOH] mol/L']
df_final = df_final.dropna(subset=['[HCl] mol/L', '[NaOH] mol/L'])

# graficar histograma de [HCl] mol/L
plt.figure(figsize=(10, 6))
sns.histplot(data=df_final, x='[HCl] mol/L', kde=True, color='#4C72B0', bins=12)
mean_hcl = df_final['[HCl] mol/L'].mean()
std_hcl = df_final['[HCl] mol/L'].std()
plt.axvline(mean_hcl, color='red', linestyle='--', label=f'Media: {mean_hcl:.3f} M')
plt.axvline(mean_hcl + std_hcl, color='gray', linestyle=':', label='1 Desviacin Estndar')
plt.axvline(mean_hcl - std_hcl, color='gray', linestyle=':')
plt.title('Distribucin de [HCl] en todos los grupos', fontsize=14)
plt.xlabel('Concentracin (M)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()
plt.show()
```

```python
# graficar el segundo histograma. concentracin de NaOH

plt.figure(figsize=(10, 6))
sns.histplot(data=df_final, x='[NaOH] mol/L', kde=True, color='#55A868', bins=12)
mean_naoh = df_final['[NaOH] mol/L'].mean()
std_naoh = df_final['[NaOH] mol/L'].std()
plt.axvline(mean_naoh, color='red', linestyle='--', label=f'Media: {mean_naoh:.3f} M')
plt.axvline(mean_naoh + std_naoh, color='gray', linestyle=':', label='1 Desviacin Estndar')
plt.axvline(mean_naoh - std_naoh, color='gray', linestyle=':')
plt.title('Distribucin de [NaOH] en todos los grupos', fontsize=14)
plt.xlabel('Concentracin (M)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()
plt.show()
```
