```python
# read the csv 'Salary_dataset.csv'
# columns id,YearsExperience,Salary 

import pandas as pd

row_dataset = pd.read_csv('Salary_dataset.csv', index_col=0)

# print the first 5 rows of the dataset
print(row_dataset.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# graph a linnear regression with equation in the plot
%pip install statsmodels

import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
# drop rows with 'YearsExperience' is over 19
row_dataset = row_dataset[row_dataset['YearsExperience'] <= 19]
```

```python

# fit the linear regression model
X = row_dataset['YearsExperience'].values.reshape(-1, 1)
y = row_dataset['Salary'].values.reshape(-1, 1)

model = LinearRegression()
model.fit(X, y)

# show the coefficients
print('Coefficient:', model.coef_[0][0])
print('Intercept:', model.intercept_[0])

```

```python
# set the figure size
plt.figure(figsize=(10, 6))

# set the title of the plot
plt.title('Salary vs Years of Experience')

# set the x and y labels
plt.xlabel('Years of Experience')
plt.ylabel('Salary')


# set the x and y ticks
plt.xticks(np.arange(0, np.max(row_dataset['YearsExperience']) + 1, 1))
plt.yticks(np.arange(0, np.max(row_dataset['Salary']) + 10000, 100000))
# set the grid
plt.grid(True)

# set the color of the grid
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# set the color of the ticks
plt.tick_params(axis='x', colors='gray')
plt.tick_params(axis='y', colors='gray')

# set the color of the labels
plt.xlabel('Years of Experience', color='gray')
plt.ylabel('Salary', color='gray')

# set the color of the title
plt.title('Salary vs Years of Experience', color='gray')

# set the color of the figure
plt.gcf().patch.set_facecolor('white')
# data points
plt.scatter(row_dataset['YearsExperience'], row_dataset['Salary'], color='blue', alpha=0.5, label='Data Points')
```
