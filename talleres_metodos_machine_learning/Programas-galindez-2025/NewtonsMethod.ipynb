```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = formula(x) 
    plt.plot(x, y)  
    #plt.show()
```

```python
graph(lambda x: 1 + x, np.arange(-3,3,0.1))
graph(lambda x: np.exp(x), np.arange(-3,3,0.1))
```

```python
graph(lambda x: 1 + x + 0.5*x**2, np.arange(-3,3,0.1))
graph(lambda x: np.exp(x), np.arange(-3,3,0.1))
```

```python
graph(lambda x: 1 + x + 0.5*x**2 + (1./6)*x**3, np.arange(-3,3,0.1))
graph(lambda x: np.exp(x), np.arange(-3,3,0.1))
```

```python
def dx(f, x):
    return abs(0-f(x))
 
def newtons_method(f, df, x0, epsilon):
    print ('x0:', x0)
    delta = dx(f, x0)
    while delta > epsilon:
        x0 = x0 - f(x0)/df(x0)
        delta = dx(f, x0)
    print ('Root at: ', x0)
    print ('f(x) at root: ', f(x0))
    return delta
```

```python
def f(x):
    return 6*x**5-5*x**4-4*x**3+3*x**2
 
def df(x):
    return 30*x**4-20*x**3-12*x**2+6*x
```

```python
graph(lambda x: 6*x**5-5*x**4-4*x**3+3*x**2, np.arange(-0.80,1.15,0.01))
```

```python
x0s = [0.25, -.65, 0.9, 1.5]
for x0 in x0s:
    newtons_method(f, df, x0, 1e-5)
```
