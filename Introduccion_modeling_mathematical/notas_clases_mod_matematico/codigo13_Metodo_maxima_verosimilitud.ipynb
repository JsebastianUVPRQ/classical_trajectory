### 1.Repaso distribuciones discretas

- **Distribución uniforme**:

La distribución uniforme  surge de la suposición de que cada evento posible tiene la misma probabilidad de ocurrir.

> - Parámetro: $N =$ número de posibles resultados
> - Espacio muestral: $\Omega = \{n_1,n_2,n_3,\dots,n_N\}$
> - Masa de probabilidad: $$P_{\rm uni}(X=m) = \frac{1}{N},\hspace{4ex} m\in \Omega$$
> - En python:
  <pre>from scipy.stats import randint
  probabilidad = randint.pmf(m, n_1, n_N)
  </pre>
  Nota: $\texttt{pmf}=$ *probability mass distribution*

- **Distribución binomial**

La distribución binomial se aplica  a  experimentos   con resultados binarios. Permite evaluar la probabilidad de obtener un cierto número de eventos favorables ("éxitos") en un número específico de repeticiones.

> - Parámetros: $n = $ número de repeticiones independientes, $p = $ probabilidad de "éxito"

> - Espacio muestral: $\Omega = \{0,1,2,3,\dots,n\}$

> - Masa de probabilidad: $$P_{\rm bin}(X=m) = \left(
  \begin{array}
   .n \\ m
  \end{array}\right)\, p^m \,(1-p)^{n-m} = \frac{n!}{m!\,(n-m)!}\, p^m \,(1-p)^{n-m} $$

> - $\hat{p} = \bar{x}/n$, donde $\bar{x}$ es la media muestral de las variables aleatorias y $\hat{p}$ es el valor estimado del parametro por el metodo de maxima verosimilitud.

> - En python:
  <pre>from scipy.stats import binom
  probabilidad = binom.pmf(m, n, p)
  </pre>


- **Distribución de Poisson**

La distribución de Poisson es una distribución discreta que permite establecer la probabilidad de que ocurra un número dado de eventos en un período de tiempo específico.

> - Parámetro: $\lambda =$ número medio de eventos que ocurren en una unidad de tiempo

> - Espacio muestral: $\Omega = \{0,1,2,3,\dots \}$

> - Masa de probabilidad: $$P_{\rm poi}(X=m) = \frac{{\rm e}^{-\lambda} \lambda^{m}}{m!}$$

> - $\hat{\lambda} = \bar{x}$, donde $\bar{x}$ es la media muestral de las variables aleatorias y $\hat{\lambda}$ es el valor estimado del parametro por el metodo de maxima verosimilitud.

> - En python:
  <pre>from scipy.stats import poisson
  probabilidad = poisson.pmf(m, lambda)
  </pre>

### 1. Ejemplo distribución binomial

Suponga que se quiere estudiar una poblacion de mariposas y de esta,  la proporción que presenta alas de color verde. Para esto se toman $N=10$ muestras independientes, cada una de ellas siendo el resultado de observar grupos de $n=20$ individuos. En cada muestra se cuenta el número de  mariposas con alas de color verde ($m$), cantidad que  corresponde a la variable aleatoria de interés.

Los resultados para $m$ de las $N$ muestras seleccionadas son los siguientes:

|   Muestra   | 'Éxitos' ($m$) | 'Fracasos' ($n-m$) |
|:-----------:|:------:|:------:|
|  Muestra 1  |   12   | 8      |
|  Muestra 2  |   9    | 11     |
|  Muestra 3  |   11   | 9      |
|  Muestra 4  |   14   | 6      |
|  Muestra 5  |   10   | 10     |
|  Muestra 6  |   13   | 7      |
|  Muestra 7  |   8    | 12     |
|  Muestra 8  |   10   | 10     |
|  Muestra 9  |   11   | 9      |
| Muestra 10  |   15   | 5      |






**Problema:**

* Estime la probabilidad de éxito $p$ de observar una mariposa con alas verdes.

* Calcule la probabilidad de observar $m=16$, $m=18$ y $m=20$ mariposas con alas verdes en una muestra de $n=20$ mariposas.

* Realice un gráfico en donde se muestre el número de mariposas de alas de color verde más probable de observar en una muestra de 20 mariposas.

**Solución:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom   # Herramientas estadísticas en el módulo scipy
```

- Estimación de la probabilidad $p$:

  - $p = $ probabilidad de que una mariposa tenga las alas verdes

  - $q=1-p = $ probabilidad de que una mariposa NO tenga las alas verdes

```python

#numero_de_observanciones_por_Muestra
n= 20


muestra_exitos = [12, 9, 11, 14, 10, 13, 8, 10, 11, 15] # Mariposas con alas verdes en cada muestra - m

# Cálculo de la media muestral
# Valor promedio de éxitos en todos los experimentos
media_muestral = np.mean(muestra_exitos)



# Estimación del parámetro p utilizando la media muestral
p_estimado = media_muestral / n

print("Media muestral = ", media_muestral, 'mariposas de alas verdes en 20 mariposas')
print("Parámetro estimado p = ", round(p_estimado,4))
```

-  Cálculo de la probabilidad de observar $m=16$, $m=18$ y $m=20$ mariposas con alas verdes en una muestra de $n=20$ mariposas

```python
# Cálculo de la probabilidad para otros eventos
valor_m1 = 10
valor_m2 = 15
valor_m3 = 20

# Usando la función de masa de probabilidad binomial
# "binom.pmf(m, n, p)" del módulo scipy
probabilidad_m1 = binom.pmf(valor_m1, n, p_estimado)
probabilidad_m2 = binom.pmf(valor_m2, n, p_estimado)
probabilidad_m3 = binom.pmf(valor_m3, n, p_estimado)

# Imprimir los resultados
print("Probabilidad de observar ", valor_m1, " mariposas de alas verdes = %8.5f\n" %probabilidad_m1)
print("Probabilidad de observar ", valor_m2, " mariposas de alas verdes = %8.5f\n" %probabilidad_m2)
print("Probabilidad de observar ", valor_m3, " mariposas de alas verdes = %8.5f\n" %probabilidad_m3)
```

- Gráfico


```python
mariposas_alas_verdes = np.arange(n + 1)
probabilidades = binom.pmf(mariposas_alas_verdes,n, p_estimado)

plt.bar(mariposas_alas_verdes, 100*probabilidades, color='red', label='$n=$'+str(n) )
plt.ylabel("Probabilidad (%))")
plt.xlabel("Número de mariposas con alas verdes")
plt.xlim([0, n])
plt.xticks(range(0,n+1,1))
plt.legend()
plt.grid(linestyle=':')
plt.show()

# Por lo tanto, observamos que el valor mas probable es de
# 11 mariposas con alas verdes por cada 20 muestreadas
```

**Comentarios:**

- *Comentario 1:* En un curso de **Estadística I** se puede demostrar que el valor mas probable siempre corresponde al valor de la media muestral.

- *Comentario 2:* En un curso de **Estadística II** se puede demostrar que a mayor número de muestras, la parte superior de las barras se aproximan a una curva del tipo $f(x) = a\,{\rm e}^{-b\,x^{2}} $, con $a$ y $b$ parametros reales. Esta curva se suele conocer como la distribucion normal, y el nombre del teorema que establece este resultado se conoce como *teorema del limite central*.

### 2. Ejemplo de distribución de Poisson

Suponga que se realizan $N=10$ experimentos para estudiar la tasa de reacciones químicas en una solución. Para ello se registró el número de reacciones que ocurrieron en cada uno de los experimentos durante 1 minuto y se obtuvieron los siguientes resultados:



|   Experimentos   | Éxitos ($m$) |
|:-----------:|:------:|
|  Experimento 1  |   20  |
|  Experimento 2  |   40   |
|  Experimento 3  |   30   |
|  Experimento 4  |   60   |
|  Experimento 5  |   50   |
|  Experimento 6  |   20   |
|  Experimento 7  |   10   |
|  Experimento 8  |   30  |
|  Experimento 9  |   40   |
| Experimento  10 |   20   |


**Problema:**

A partir de los resultados obtenidos:

* Encuentre la tasa media de reacciones por minuto, denotada como $\lambda$.

* Encuentre la probabilidad porcentual de que ocurran $m=25$ reacciones en un minuto.

* Encuentre la probabilidad porcentual de que ocurran $m=1918$ reacciones en una hora.

* Realice un gráfico en donde se muestre el número de reacciones más probable en 2 minutos en un intervalo de reacciones entre $10$ y $100$.

**Solución:**

```python
import numpy as np
from scipy.stats import poisson
```

- Evaluación de $\lambda$

```python
# Datos de muestra
muestra = [20, 40, 30, 60, 50, 20, 10, 30, 40, 20]

# Estimación del parámetro lambda usando la media muestral
lambda_estimado = np.mean(muestra)

# Imprimir los resultados
print("Parámetro estimado lambda:", lambda_estimado, 'reacciones/min')
```

- Probabilidad de que ocurran $m=25$ reacciones en un minuto

```python
# Cálculo de la probabilidad para obtener 25 reacciones quimicas
valor_m1 = 25
probabilidad_m1 = poisson.pmf(valor_m1, lambda_estimado)

print("Probabilidad porcentual de obtener", valor_m1, "reacciones en un minuto = ", round(100*probabilidad_m1,2), "%")
```

- Probabilidad de obtener $m=1918$ reacciones en una hora:



> Aquí se debe cambiar la escala de tiempo asociada al parámetro $\lambda$ de  $\frac{\rm reacciones}{\rm minuto}$ a  $\frac{\rm reacciones}{\rm hora}$.


> Para ello, se debe multiplicar a $\lambda_{\rm estimado}$ por $60$. Después de esto, se puede aplicar  la distribucion de Poisson para estimar la probabilidad requerida.

```python
# Cálculo de la probabilidad para obtener 1918 reacciones en una hora
valor_m2 = 1918

lambda_estimado_hora = 60*lambda_estimado
probabilidad_m2 = poisson.pmf(valor_m2, lambda_estimado_hora)

print("Probabilidad porcentual para obtener", valor_m2, "reacciones en una hora = ", round(100*probabilidad_m2,2), "%")
```

- Gráfico

> Al igual que antes, se debe cambiar la escala de tiempo asociada al parámetro $\lambda$ de  $\frac{\rm reacciones}{\rm minuto}$ a  $\frac{\rm reacciones}{\rm dos~minutos}$.

```python
lambda_estimado_dos = 2*lambda_estimado

numero_reacciones = np.arange(10,101)
probabilidades = 100*poisson.pmf(numero_reacciones, lambda_estimado_dos)

plt.bar(numero_reacciones, probabilidades, color='green' )
plt.ylabel("Probabilidad (%)")
plt.xlabel("Reacciones por cada dos minutos")
plt.xlim([30, 100])
plt.xticks(range(30,101,10))
plt.grid(linestyle=':')
plt.show()

# Por lo tanto, observamos que el valor mas probable es
# de alrededor de 64 reacciones en dos minutos
```
