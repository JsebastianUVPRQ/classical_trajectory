## 2.1 Estructura del archivo fuente

1. **Preámbulo (encabezado)**

  - Información sobre los aspectos globales del documento
    - Tipo (*paper, report, book, slides, letter*)
    - Tamaño de papel y de letra
  - Herramientas que se deben cargar
    - Paquetes (*amsmath, babel, graphicx, inputenc, etc*)
 - Comienza con la declaración del tipo  de documento:
 <pre>\documentclass[opciones]{tipo de documento}</pre>

2. **Cuerpo**

  - Contenido del documento junto con las instrucciones (comandos) para darle formato
  - Se encuentra encerrado por los comandos:
  <pre>\begin{document}
  ...
  ...
\end{document}</pre>

3. **Instrucciones**

  - Comandos
    - Son órdenes que sirven para que LaTeX realice una acción “sencilla”
    - Ejemplo: cambiar el formato, color o tamaño del texto
    - La sintaxis habitual es:
    <pre>\comando[opciones]{argumento}</pre>
  
  - Entornos
    - Son órdenes que sirven para que LaTeX realice una acción “compleja”
    - Ejemplo: crear una tabla, insertar una figura
    - La sintaxis es:
  <pre>\begin{entorno}
  ...
  ...
\end{entorno}</pre>
