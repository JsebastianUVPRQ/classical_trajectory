## 1. Cálculo de probabilidades

### 1.1 Enfoque frecuentista

De acuerdo con este enfoque, la probabilidad de ocurrencia de un evento se calcula mediante la observacion y el análisis empírico de los resultados de la repetición del experimento.

**Problema:**

Supongamos que queremos lanzar un dado (*experimento*). Entonces  el [*espacio muestral*](https://https://es.khanacademy.org/math/4-grado-innova-schools/x4d4e13c9a2690da8:incertidumbre-y-datos/x4d4e13c9a2690da8:espacio-muestral/v/espacio-muestral)  es $\Omega = \{1, 2, 3, 4, 5, 6\}$, donde cada número representa uno de todos los resultados posibles al lanzar el dado.

- Evaluar las probabilidades de obtener los siguientes resultados tras un lanzamiento:
 - Probabilidad de obtener un seis.
 - Probabilidad de obtener un número par.
 - Probabilidad de obtener un número mayor que $4$.

- Realizar un histograma que muestre la frecuencia de ocurrencia de los resultados obtenidos.


**Solución:**

- Selección de un "número aleatorio"

```python
# Módulo para trabajar con variables aleatorias
import random as rnd

# Función randint(N_ini, N_fin)
#   Escoge aleatoriamente un número entero M tal que N_ini <= M <= N_fin
numero_aleatorio = rnd.randint(1,10)
print(numero_aleatorio)


# Función choice(secuencia_elementos)
#   Escoge aleatoriamente un elemento de la secuencia dada
secuencia_elementos = [1, 2, 3, 'home']
print( rnd.choice(secuencia_elementos) )
```

- Cálculo de las probabilidades

```python
import random as rnd
import matplotlib.pyplot as plt
```

```python
# Función para simular un lanzamiento de un dado
def lanzamiento_un_dado():
  # Arroja un número escogido aleatoriamente
  # del conjunto {1, 2, 3, 4, 5, 6}
  # 1ra opción
  resultado = rnd.randint(1,6)
  # 2da opción
  #resultado = rnd.choice([1,2,3,4,5,6]])
  return resultado
```

```python
# Simula una serie de lanzamientos del dado
# ("repeticiones del experimento") y guarda
# el resultado de cada uno

numero_lanzamientos = 1000   # Número de repeticiones

lista_resultados = []
for n in range(numero_lanzamientos):
    # "Lanzamiento de un dado"
    resultado_lanzamiento =lanzamiento_un_dado()
    # Agrega (concatena) el resultado obtenido a la lista de resultados
    lista_resultados.append(resultado_lanzamiento)

print('Resultados = ', lista_resultados[0:12], '...')
```

```python
# Calcula la probabilidad de obtener un seis

numero_veces_seis = 0
for resultado in lista_resultados:
  # Revisa entre todos los resultados los que
  # arrojaron el "evento de interés"
  if resultado == 6:
    # Cuenta las veces que se obtuvo el resultado de interés (frecuencia)
    numero_veces_seis += 1

# Evalúa la frecuencia relativa (probabilidad frecuentista) del evento de interés
probabilidad_seis = numero_veces_seis / numero_lanzamientos

# Imprime los resultados para las frecuencias relativas
# (probabilidad frecuentista) de los eventos de interés
print('Probabilidad de obtener ...')
print('... un seis =', probabilidad_seis)
```

```python
# Calcula la probabilidad de obtener un número par

numero_veces_par = 0
for resultado in lista_resultados:
  # Conteo de frecuencia
  if resultado % 2 == 0:   # A%B evalúa el módulo
    numero_veces_par += 1

# Evalúa la frecuencia relativa del evento de interés
probabilidad_par = numero_veces_par / numero_lanzamientos

# Imprime los resultados para las frecuencias relativas
# (probabilidad frecuentista) de los eventos de interés
print('Probabilidad de obtener ...')
print('... un número par =', probabilidad_par)
```

```python
# Calcula la probabilidad de obtener un número mayor que 4

numero_veces_mayor_4 = 0
for resultado in lista_resultados:
  # Conteo de frecuencia
  if resultado > 4:
    numero_veces_mayor_4 += 1

# Evalúa la frecuencia relativa del evento de interés
probabilidad_mayor_4 = numero_veces_mayor_4 / numero_lanzamientos

# Imprime los resultados para las frecuencias relativas
# (probabilidad frecuentista) de los eventos de interés
print('Probabilidad de obtener ...')
print('... un número mayor que 4 =', probabilidad_mayor_4)
```

> Un [histograma](https://es.wikipedia.org/wiki/Histograma) es una representación gráfica de la distribución de datos numéricos, en la que cada barra indica el número de veces (*frecuencia*) que un resultado se encuentra en un rango determinado

```python
# Crea un histograma a partir de los resultados de las
# repeticiones del experimento de lanzar un dado

plt.hist(lista_resultados, bins=range(1, 8), align='left', rwidth=0.8, density=False)
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.title('Histograma de lanzamientos de un dado')
plt.xticks(range(1,7))
plt.grid(linestyle=':')
plt.show()
```

### 1.2 Enfoque clásico

De acuerdo con este enfoque:

- Probabilidad de que ocurra el evento A "Obtener un seis"

  $$P(A) = P(\{6\}) = \frac{1}{6} \simeq  0.1667$$

> La probabilidad de obtener un seis es $1/6$, ya que hay un resultado favorable ($6$) de un total de seis posibles resultados

- Probabilidad de que ocurra el evento B "Obtener un número par"

  $$P(B) = P(\{2, 4, 6\}) = \frac{3}{6} = \frac{1}{2} =  0.5$$

>  La probabilidad de obtener un número par es $1/2$, ya que hay tres resultados favorables ($2, 4$ y $6$) de un total de seis posibles resultados.

- Probabilidad de que ocurra el evento C "Obtener un número mayor que 4"

  $$P(C) = P(\{5, 6\}) = \frac{2}{6} \simeq 0.3333 $$


> La probabilidad de obtener un número mayor que $4$ es $\frac{1}{3}$, ya que hay dos resultados favorables ($5$ y $6$) de un total de seis posibles resultados.

### 1.3 Ejemplo: Lanzamiento de una moneda

Analizar a través del enfoque frecuentista la probabilidad de obtener 'cara' $(P_{\rm cara})$ en el lanzamiento de una moneda. Para ello, (i) evalúe dicha probabilidad cambiando el número de repeticiones $(N_{\rm rep})$ del experimento y (ii) realice un gráfico $P_{\rm cara}$ vs $N_{\rm rep}$, en el cual se observe que a mayor número de lanzamientos, la probabilidad de obtener "cara" se aproxima a $1/2 = 0.5$, como lo indica el enfoque clásico.

**Solución**

```python
import random as rnd
import matplotlib.pyplot as plt
```

```python
# Función para simular el lanzamiento de una moneda
def lanzar_moneda():
    resultado = rnd.choice(['cara', 'sello'])
    return resultado

# Probabilidad clásica de obtener cara
probabilidad_clasica = 0.5  # (1 evento favorable) / (un total de 2 eventos posibles)
```

```python
# Se realizan N repeticiones del experimento
# y la probabilidad se va calculando para cada
# repetición

# Número máximo de repeticiones del experimento
numero_repeticiones_max = 5000

# Lista para almacenar la probabilidad frecuentista
# para cada número de repeticiones del experimento
numero_repeticiones = list(range(1, numero_repeticiones_max+1))
probabilidad_frecuentista = []

# Simula N lanzamientos de una moneda), guarda los
# resultados y analiza la probabilidad frecuentista
lista_resultados=[]
for N_rep in numero_repeticiones:
    # Realiza la jugada y acumula el resultado
    resultado_lanzamiento = lanzar_moneda()
    lista_resultados.append( resultado_lanzamiento )

    # Evalua la probabilidad frecuentista hasta la repetición actual
    frecuencia_cara   = lista_resultados.count('cara')
    probabilidad_cara = frecuencia_cara / N_rep
    probabilidad_frecuentista.append( probabilidad_cara )

print('Resultados experimento = ', lista_resultados[0:6],'...\n')
print('Probabilidad de obtener "cara" = ', probabilidad_cara , 'para', N_rep, 'repeticiones\n' )
```

```python
# Crear el gráfico
plt.plot(numero_repeticiones, probabilidad_frecuentista, '*:', label='Probabilidad frecuentista')
plt.axhline(y=probabilidad_clasica, color='r', linestyle='-', label='Probabilidad clásica')
plt.title('Comparación de probabilidades frecuentista y clásica')
plt.xlabel('Número de repeticiones')
plt.ylabel('Probabilidad de obtener cara')
plt.legend()
plt.grid()
plt.show()
```
