De acuerdo con la fisica newtoniana, el movimiento de un [péndulo simple](https://es.wikipedia.org/wiki/P%C3%A9ndulo_simple) tomando en cuenta la fricción del aire  se puede describir a través de la ecuación diferencial

\begin{equation}
\dfrac{d^2 \theta}{d t^2} = - \dfrac{g}{l}\,\sin\theta - b \,\omega ,
\end{equation}

donde $\theta$ (*theta*) representa el ángulo con la vertical, $\omega=\frac{d\theta}{d t}$ (*omega*) la velocidad angular, $g$ la acelaración de la gravedad, $l$ la longitud del péndulo y $b$ la intensidad del amortiguamiento debido al aire.

```python
# Definición de la EDO
import numpy as np
import matplotlib.pyplot as plt
# Definimos la EDO

# Parámetros del sistema
L = 2.0    # m
g = 9.8    # m/s^2
b = 0.04   # 1/s

# EDO
def pendulo(theta, omega):
  dtheta_dt = omega
  domega_dt = -(g/L)*np.sin(theta) - b*omega
  return dtheta_dt, domega_dt


```
