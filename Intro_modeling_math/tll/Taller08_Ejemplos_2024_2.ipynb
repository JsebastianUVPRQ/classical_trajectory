## 1. Repaso de derivadas

La [derivada](https://https://www.geogebra.org/m/P5DRE8R4)  de la función $f(x): \mathbb{R}\to\mathbb{R} $ es una medida del cambio que ocurre en la cantidad $y=f(x)$ cuando se realiza un pequeño cambio en la cantidad $x$.  Formalmente, la derivada se define a través del límite

$$f^\prime(x)= \frac{df}{dx} = \lim_{h \to 0}\frac{f(x+h)-f(x)}{h},$$

el cual se puede interpretar geométricamente  como la pendiente de la recta tangente a la función en el punto $x$, como se indica el trazo rojo en la figura.

<img src=https://www.math.net/img/a/calculus/limit-definition-of-a-derivative-1.png  width=400>

desde el punto de vista del modelamiento, su importancia radica en que
- muchas entidades físicas se definen como la tasa instantánea de cambio de alguna otra cantidad
  - ejemplo: la velocidad se relaciona con la derivada temporal del desplazamiento
- la derivada permite "optimizar" una función
  - ejemplo: encontrar el máximo o el mínimo de una función

## 2. Ecuaciones diferenciales ordinarias

Una ecuacion diferencial ordinaria (EDO) es una ecuación en las que se relacionan una variable independiente $x$, una funcion $y(x)$, que es desconocida, y las derivadas de $y(x)$. Matemáticamente

$$\mathcal{F}\left[x, y, \frac{dy}{dx},\cdots, \frac{d^ny}{dx^n}\right] = 0,$$

donde $n$ es llamado el orden de la ecuación.

Este tipo de relaciones entre una cantidad y sus tasas de cambio son frecuentes en áreas de las ciencias naturales, la ingeniería o la economía. Existe una gran cantidad de fenómenos reales complejos que pueden ser formulados en términos de ecuaciones diferenciales. En particular, algunos *sistemas dinámicos en tiempo continuo* pueden modelarse mediante  EDOs, lo cual permite predecir sus comportamientos futuros, entre otras cosas.

En los problemas a tratar en este curso nos limitaremos a problemas de primer $(n=1)$ y segundo orden $(n=2)$.








**Ejemplo: Ley de enfriamiento/calentamiento de Newton**

La ley de enfriamiento/calentamiento de Newton establece que la rapidez con la que cambia la temperatura de un objeto es proporcional a la diferencia entre su temperatura y la del medio que lo rodea ( temperatura del ambiente). Si $T(t)$ indica la temperatura del cuerpo en el tiempo $t$ y $T_m$ es la temperatura del ambiente (que se considera constante), entonces, la ley de  enfriamiento/calentamiento de Newton determina que

$$\frac{dT}{dt} = -k \left( T  - T_m \right),$$

donde $k$ es una constante de proporcionalidad relacionada con la capacidad de transmisión de calor del objeto al medio y al área superficial del objeto.


### 2.1 Problema de valor inicial

Dada una EDO de orden $n$, un intervalo $I$ y un punto $x_0 \in I$, se llama *problema de valor
inicial* al problema de encontrar la función $y(x)$ que satisface la EDO en el intervalo $I$ y que además  satisface todas las condiciones iniciales

$$y(x_0)=c_0,\quad y^\prime(x_0)=c_1, \dots,\quad  y^{(n-1)}(x_0)=c_{n-1} $$

**Ejemplo:**

Suponga que un objeto que tiene una temperatura de 5°C es puesto (en el tiempo $t=0$) en un lugar donde la temperatura se mantiene a 25°C y $k=2\times 10^{-3}~\text{s}^{-1}$. ¿Cuál será la temperatura del objeto después de 5 minutos?

El problema de valor inicial a resolver es entonces:
$$
\begin{cases}
\dfrac{dT(t)}{dt} &=& -0.12 \left[T(t)-25\right]\\\\
T(0) &=& 5
\end{cases}
$$
con $T$ medida en grados Celsius y $t$ en minutos

## 3. Solución numérica de EDOs: Método de Euler

Suponga que queremos resolver numéricamente el problema de valor inicial

$$\dfrac{d y(t)}{dt} = f(t,y),\qquad  y(t_0) = y_0.$$

en el intervalo  $t_0\le t \le t_{\rm max}$. El [método de Euler](https://es.wikipedia.org/wiki/M%C3%A9todo_de_Euler) es el método  numérico  más sencillo para solucionar un problema de valor inicial.

Este método se basa en aproximar la definición de derivada para un cambio $t \to t+\Delta t$ suficientemente pequeño de la variable independiente como:

$$\frac{dy}{dt} \approx \frac{y(t+\Delta t)-y(t)}{\Delta t}.$$






Utilizando el método de Euler se puede aproximar la función $y_i = y(t_i)$ evaluándola en una sucesión de tiempos $t_{i}$ $(i=0, 1, \dots,N)$ mediante la formula de recurrencia (*diferencias finitas*)

$$ y_{i+1} = y_{i} + f(t_i,y_i)\,\Delta t, $$

donde $\Delta t = t_{i+1} - t_{i}$ (paso de integración), $t_0$ es el tiempo inicial y $t_N$ es el tiempo máximo $\left(N = \left\lfloor t_{\rm max}/\Delta t  \right\rfloor\right)$.  El método se esquematiza en el correspondiente [diagrama de flujo](https://drive.google.com/file/d/1C9jaD-8vHEZW9XeZSyAcIgUHyHM9KUeb/view?usp=drive_link).

> **Nota**: Dadas sus aproximaciones, el método de Euler tiene una precisión limitada, que decrece a medida que $\delta t$ crece. Este efecto es aún mayor si las funciones involucradas decaen u oscilan muy rápidamente. Existen otros métodos para solucionar EDO que corrigen las limitaciones del método y garantizan mayor precisión.

### 3.1 Ejemplo

Suponga que un recipiente con agua a una temperatura $T_0 = 8\text{°C}$ es puesto en el tiempo $t=0$ en un lugar donde la temperatura del ambiente se mantiene a 25°C. Considere además que para el recipiente dado, la tasa de enfriamiento/calentamiento tiene un valor constante $k=1.45\times 10^{-3}~\text{s}^{-1}$.
- Use el método de Euler para resolver el problema de valor inicial para $0\le t \le 90$ minutos
- ¿Cuál será la temperatura del recipiente después de 30 minutos?  
- Este problema se puede resolver exactamente, siendo la solución

  $$T_{\rm exacta} (t) = T_0 + (T_m-T_0)\,\left[\,1 - {\rm e}^{-k\cdot t}\,\right] $$

  Grafique la diferencia porcentual entre el resultado exacto y el numérico, es decir,
  
  $$\Delta T(\%)  = 100\,\left|\frac{T(t)-T_{\rm exacta}(t)}{T_{\rm exacta}(t)}\right| $$


```python
# Importa las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
# Define el estilo y tamaño de las gráficas
# plt.style.use ('default')
# plt.rcParams['figure.figsize'] = [7.0, 4.0]
```

* Definir la función $\dfrac{dT}{dt}=f(T)$ necesaria

```python
# Definimos la función f(T)

# Parámetros del sistema
k_s = 0.00145   # 1/s
k_m = 60*k_s   # 1/min
Tm = 25        # °C

# Función f(T) -  Derivada
def derivada(T):
  # Temperatura en °C
  # Tiempo en minutos
  dT_dt = -k_m * (T - Tm)
  return dT_dt
```

* Definir los arreglos necesarios

```python
# Definimos parámetros
T0   = 8     # T inicial (°C)
tmax = 90    # t maximo (min)
dt   = 0.6  # Delta_t (min)

# Definimos los arreglos necesarios
tiempo = np.arange(0,tmax+dt,dt)      # Tiempo
Temperatura = np.zeros_like(tiempo)   # Temperatura
Temperatura[0] = T0                   # Initial condition
```

* Evaluar la recurrencia  $T_{i+1} = T_{i} + f(t_i,y_i) \, dt$



```python
# Método de Euler
Npoints = len(tiempo)

i=0
while i < (Npoints-1):
    dT_dt = derivada(Temperatura[i])
    # T[i+1] = T[i] + Cambio_de_T
    Temperatura[i+1] = Temperatura[i] + (dT_dt * dt)
    i += 1

plt.plot(tiempo,Temperatura,label='dt='+str(dt))
plt.xlabel('$t$ (min)',size=12)
plt.ylabel('$T$ (°C)',size=12)
plt.grid()
plt.axis([0,tmax,T0,Tm])
plt.legend()
plt.show()
```

```python
# Npoints --> 90 min
#     N30 --> 30 min
N30 = round( Npoints*30/90)
print('Para t = ', tiempo[N30], 'min la temperatura del agua es', round(Temperatura[N30],2), '°C')
```

* Calcular las diferencias

```python
# Solución exacta
def Temperatura_exacta(t):
  # Temperatura en °C
  # Tiempo en minutos
  Tex = T0 + (Tm-T0)*(1-np.exp(-k_m*t))
  return Tex

# Evalúa las diferencias
diferencia = 100*np.abs((Temperatura-Temperatura_exacta(tiempo))/Temperatura_exacta(tiempo))

# Grafica la diferencia porcentual
plt.plot(tiempo,diferencia,label='dt='+str(dt))
plt.xlabel('$t$ (min)')
plt.ylabel('$\Delta T(\%)$')
plt.grid()
plt.legend()
plt.show()
```
