###Ejemplo (probabilidad frecuentista)

Supongamos que queremos lanzar un dado. Entonces  el espacio muestral es $\Omega = \{1, 2, 3, 4, 5, 6\},$ donde cada número representa un resultado posible al lanzar el dado. Calculemos las siguientes probabilidades


- Probabilidad de obtener un seis.
- Probabilidad de obtener un número par.
- Probabilidad de obtener un número mayor a $4$.

Verifique que las probabilidades obtenidas se aproximan a los valores obtenidos por el enfoque clásico si realiza un número de lanzamientos "grande". Adicionalmente, realice un histograma que muestre los resultados obtenidos.


**Solucion**

Como generar numeros aleatorios?

**Respuesta**: Usando la funcion de numpy random.rand() y random.randint()

```python
#usando el modulo numpy
import numpy as np


#Genreacion de numeros entre 0 y 1 (por defecto para np.random.rand)
print(np.random.rand())


#Tambien podemos generar un arreglo unidimensional
print(np.random.rand(3))


#Tambien podemos generar un arreglo matricial!
#print(np.random.rand(3,2))


#generar numeros aleatorios enteros!
print( np.random.randint(1,6)  )



#Bonu: generar un arreglo de numeros aleatorios enteros!
#print(  np.random.randint(1,5, size=(2, 2) )  )

```

Con la funcion rand del modulo random, podemos simular el lanzamiento de un dado de la siguiente manera:

```python
import random  #Modulo para generar numeros aleatorios
import matplotlib.pyplot as plt


# Función para simular el lanzamiento de un dado
def lanzar_dado():
   #funcion para generar numeros aleatorios entre 1 y 6
    return random.randint(1, 6)



# Realizar una serie de lanzamientos de dado
num_lanzamientos = 10000
resultados = []


for _ in range(num_lanzamientos):
    resultado = lanzar_dado()
    resultados.append(resultado)





# Calcular la probabilidad de obtener un seis
num_seis= 0
for resultado in resultados:
    if resultado == 6:
        num_seis += 1
probabilidad_seis = num_seis / num_lanzamientos






# Calcular la probabilidad de obtener un número par
num_par = 0
for resultado in resultados:
    if resultado % 2 == 0:
        num_par += 1
probabilidad_par = num_par / num_lanzamientos




# Calcular la probabilidad de obtener un número mayor a 4
num_mayor_4 = 0
for resultado in resultados:
    if resultado > 4:
        num_mayor_4 += 1
probabilidad_mayor_4 = num_mayor_4 / num_lanzamientos




# Imprimir los resultados
print("Probabilidad de obtener un seis:", probabilidad_seis)
print("Probabilidad de obtener un número par:", probabilidad_par)
print("Probabilidad obtener número mayor a 4:", probabilidad_mayor_4)
```

- Calculo del histograma

```python

# Crear un histograma de los resultados
plt.hist(resultados,   bins = [1,2,3,4,5,6,7]  )

plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.title('Histograma de lanzamientos de un dado')

plt.grid(True)
plt.show()
```


**Explicación**

La opción bins en la función plt.hist de Matplotlib juega un papel crucial en la representación de la distribución de datos utilizando histogramas. Su función principal es especificar el número de intervalos o "contenedores" en los que se agrupan los datos para crear las barras del histograma.

Si se pasa una secuencia o un array a bins, se interpretan como los bordes de los contenedores. Esto te da un control más preciso sobre la ubicación y el ancho de cada contenedor, lo que permite personalizar la representación del histograma.

En nuestro caso al escribir bins=[1,2,3,4,5,6,7], estamos diciendo que tenemos los siguientes contenedores


- b1= [1,2]
- b2= [2,3]
- b3= [3,4]
- b4= [4,5]
- b5= [5,6]
- b6= [6,7]

Entonces, la funcion hist, toma los datos y los agrupa en estos intervalos!

- Función de histograma mas refinada!

```python

#un histograma mucho mas depurado!

plt.hist(resultados,  rwidth=0.5 , edgecolor='black', bins = [1,2,3,4,5,6,7] ,  align='left' , density=True )

plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.title('Histograma de lanzamientos de un dado')

plt.grid(True)
plt.show()
```

Calcule las probabilidades del ejemplo anterior usando el enfoque clasico.

**Solucion**


- Evento A: Obtener un seis
        
  $$ P(A) = P(\{  6\}) = \frac{1}{6} = 0.166666 $$

<br>

<br>

  $$ P(A) = P(\{2, 4, 6\}) = \frac{3}{6} = 0.5  $$
  
  La probabilidad de obtener un número par es de $\frac{1}{2}$, ya que hay tres resultados favorables $2, 4$ y $6$ en un total de seis posibles resultados.


<br>

<br>

- Evento C: Probabilidad obtener número mayor que $4$

    
  $$ P(B) = P(\{5, 6\}) = \frac{2}{6} = 0.3333 $$

### Ejemplo 3 (clasico y frecuentista)

Supongamos que queremos lanzar DOS dado. Usando el enfoque frecuentista realice un histograma que muestre el valor mas probable es obtener 7. Verifique que la probabilidad  obtenida se aproxima al valor obtenido por el enfoque clásico si realiza un número de lanzamientos "grande".

**Solución**

*Enfoque frecuentista*

```python
import random
import matplotlib.pyplot as plt

# Número de lanzamientos
num_lanzamientos = 10000

# Lista para almacenar los resultados de los lanzamientos
resultados = []

# Realizar los lanzamientos y guardar los resultados
for _ in range(num_lanzamientos):

    #Resultado = lanzar dado 1      +  Lanzar dado 2
    resultado = random.randint(1, 6) + random.randint(1, 6)
    resultados.append(resultado)


#Crear el histograma

#Con el comando  density=True obtenemos la "frecuencia relativa" en el histograma. Es decir, todos los valores del eje y se dividen por el numero total de datos!
plt.hist(resultados, bins=range(2, 14),  align='left',  rwidth=0.8, color='red', density=True)

# Configurar el gráfico
plt.title('Histograma de resultados al lanzar dos dados')
plt.xlabel('Suma de los valores')
plt.ylabel('Frecuencia relativa')


# Mostrar el gráfico
plt.show()
```

*Enfoque clasico*



 -El espacio muestral $S$ esta dado por $$S:=\{ (i,j) | i,j =1,2,3,4,5,6 \}$$


 -En este caso, el evento de sacar $7$ esta dado por el conjunto: $$A=\{(6,1),(5,2),(4,3),(3,4),(2,5),(1,6) \}$$



 - Por lo tanto:
    - $N$ es el numero total de elementos del espacio muestral: $N = 36$
    
    - $n_a$ numero de puntos que componen el evento $A$:  $n_a = 6$



- Claramente, $P(A) =  \dfrac{ n_{a} }{N} = \dfrac{6}{36} = 0.166$
